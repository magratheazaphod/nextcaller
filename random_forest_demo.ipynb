```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import roc_auc_score
import pandas as pd
```

```python
x = pd.read_csv("train.csv")
x.dropna(inplace = True)
```

```python
y = x.pop("Survived")
```

```python
model =  RandomForestRegressor(n_estimators = 1000 , oob_score = True, random_state = 42)
```

```python
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']

newdf = x.select_dtypes(include=numerics)
```

```python
newdf
```

```python
model.fit(newdf,y)
```

```python
print("AUC - ROC : ", roc_auc_score(y,model.oob_prediction_))
```

```python
sample_leaf_options = [1,5,10,50,100,200,500]
```

```python
for leaf_size in sample_leaf_options:
    model = RandomForestRegressor(n_estimators = 200, oob_score = True, n_jobs = -1,random_state =50,\
                                  max_features = "auto", min_samples_leaf = leaf_size)
    model.fit(newdf,y)
    print("AUC - ROC : ", roc_auc_score(y,model.oob_prediction_))
```
